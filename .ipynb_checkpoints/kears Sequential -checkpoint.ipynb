```python
from keras.models import Sequential 
from keras.layers import Dense,Activation
from keras.utils import to_categorical
```

```python

model = Sequential([
Dense(32,input_dim=(100)),
Activation('sigmoid'),
Dense(10),
Activation('softmax'),
])
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python

# Generate dummy data
import numpy as np
data = np.random.random((1000, 100))
labels = np.random.randint(10, size=(1000, 1))

one_hot_labels = to_categorical(labels, num_classes=10)
```

```python
model.fit(data, one_hot_labels, epochs=10, batch_size=32)
```

```python

model = Sequential()
model.add(Dense(32, activation='relu', input_dim=100))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Generate dummy data
import numpy as np
data = np.random.random((1000, 100))
labels = np.random.randint(10, size=(1000, 1))

# Convert labels to categorical one-hot encoding
one_hot_labels = to_categorical(labels, num_classes=10)

# Train the model, iterating on the data in batches of 32 samples
model.fit(data, one_hot_labels, epochs=500, batch_size=32)
```

```python
t_data= np.random.random((100,100))
t_labels=np.random.randint(10,size=(100,1))
t_labels= to_categorical(t_labels)
score = model.evaluate(t_data, t_labels, batch_size=128)
print score
```
